```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Creating a vertical bar chart
# To create a bar chart using the object-oriented, use the interface method 
 # ax.bar() function and add the x and y data parameters inside the parentheses.
    
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label='New York')
```

```python
# create a horizontal bar chart using the object-oriented 
# interface method, use the ax.barh() function.
## Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
```

```python
# SCATTER PLOTS 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# plt.plot() method 
# When we use the plt.plot() function to create a scatter plot, we need to add a lowercase o as a 
# parameter inside the parentheses. This switches the plot from a line chart to a scatter plot.
plt.plot(x_axis, y_axis, 'o')
```

```python
# plt.scatter() method
# To create a scatter plot using the plt.scatter() function, add the x-axis and y-axis parameters inside the parentheses
plt.scatter(y_axis, x_axis, c="red", label='Chicago')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Creating a Bubble Chart Using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)

# Making all points on scatter plot three times bigger. 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

# Then use the new y-axis list for the s parameter:
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Avoiding "for" and "when" loops. 
# Using the list comprehension technique to multiply each data point in the y-axis list, 
# we would do the following in the plt.scatter() function:
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c="skyblue", label='Boston', alpha=.20, edgecolor='Black', linewidth=2)
#ax.set_markeredgecolor="Black"
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Create a title.
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
# Add the legend.
ax.legend()
```

```python
# Pie Chart
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Formatting the colors in pie chart using CSS: https://matplotlib.org/3.1.1/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, explode=explode, colors=colors, shadow=True, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
